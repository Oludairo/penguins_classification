###### dropping the irrelevant columns

```python
df = df.drop(['studyName','Sample Number','Individual ID','Stage','Comments','Region','Date Egg'], axis=1)
```

###### using the ordinal encoder to convert objects to floats

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
encode = OrdinalEncoder()
df[['specie', 'island','clutch_completion','sex']] = encode.fit_transform(df[['Species', 'Island','Clutch Completion','Sex']])
```

###### dropping the older columns

```python
df_n = df.drop(columns=['Species', 'Island','Clutch Completion','Sex'])
```

```python
df_n
```

###### checking for missing values

```python
df_n.info()
```

```python
df_n.isna().sum()
```

###### filling missing values with median values

```python
df_ = df_n.fillna(df_n.median())
```

```python
df_
```

```python
df_.info()
```

###### for loop plotting box plots to check for outliers

```python
for columns in ['Culmen Length (mm)', 'Culmen Depth (mm)', 'Flipper Length (mm)', 'Body Mass (g)', 'Delta 15 N (o/oo)', 'Delta 13 C (o/oo)']:
    plt.figure(figsize=(8, 5))
    sns.boxplot(data = df_, x=columns)
    plt.title(columns)
    plt.show()
```

######  heat map showing correlation between dependent and independent variables

```python
plt.figure(figsize=(15, 10))
sns.heatmap(df_.corr(), annot=True)
```

###### splitting dataset into depedent and independent variables

```python
X = df_.drop(['specie'], axis=1)
y = df_['specie']
```

###### using the variance inflation factor to check for multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def calcu_mult(df):
  vif = pd.DataFrame()
  vif["variables"] = X.columns
  vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

  return vif
```

```python
calcu_mult(X)
```

```python
y.value_counts()
```

###### Using smote to balance our categorical dependent variable in other to avoid bias in our model

```python
from imblearn.over_sampling import SMOTE
```

```python
sample = SMOTE()
X, y = sample.fit_resample(X, y)
```

```python
y.value_counts()
```

###### normalizing dataset using the standard scaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
normalizer = StandardScaler()
norm = pd.DataFrame(normalizer.fit_transform(X))
```

###### splitting our normalized dataset into train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(norm, y, test_size=0.2, random_state=45)
```

###### using the principal component analysis to check for important features

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components= 9)
model = pca.fit(norm)
PCA(n_components= 9)
```

```python
features = X.columns
importances = model.explained_variance_ratio_
indices = np.argsort(importances)[-9:]
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

###### using the elbow method to determine the number of clusters in a data set

```python
from sklearn.cluster import KMeans
```

```python
lst = []
for ks in range(2, 20):
    km = KMeans(n_clusters = ks)
    kmeans = km.fit(X)
    lst.append(kmeans.inertia_)

fig = plt.figure(figsize= (15, 7))
plt.plot(range(2, 20), lst)
plt.grid(True)
plt.title("elbow_method")
```

###### from yellowbrick.cluster import SilhouetteVisualizer

```python
fig, ax = plt.subplots(2, 2, figsize=(15,8))
for i in [2, 3, 4, 5]:
    km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=100, random_state=42)
    q, mod = divmod(i, 2)

    visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax[q-1][mod])
    visualizer.fit(norm)
```

###### fitting the training test into K-Neighbors Classifier model

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
neigh = KNeighborsClassifier(n_neighbors=5)
knn_model = neigh.fit(X_train, y_train)
```

###### checking accuracy score

```python
knn_model.score(X_train, y_train)
```

###### predicting with the test

```python
y_pred = knn_model.predict(X_test)
```

###### using the confusion matrix to display the four quadrants that show the true and false positive and negative classes

```python
confusion_matrix(y_test, y_pred)
```

#### From the confusion matrix above the model was able to classify 28 out of 29 for Adelie Penguin species, and 32 of Chinstrap penguin, 31 Gentoo penguin printing the precision,recall,and f1-score to determine how well the model has classified the dataset.

```python
print(classification_report(y_test, y_pred))
```

#### From the above classification report we can see that our model has 0.98 percent for classifying correctly the Adelie Penguin, 0.98 percent for classifying correctly the Chinstrap penguin, and then 100 percent for classifying correctly the Gentoo penguin with accuracy of 0.99.
#### using the repeated k fold model to cross validate

```python
from sklearn.model_selection import RepeatedKFold, cross_val_score
```

```python
Cv_ = RepeatedKFold(n_splits=10, n_repeats=4, random_state=1)
score = cross_val_score(knn_model, X_test, y_test, scoring = "accuracy", cv = Cv_, n_jobs=1, error_score="raise")
np.mean(score)
```

####  The cross validation of our model shows that our model has 0.99 percentage of predicting correctly which is a good score given that our train and test score is the same
